```python
# el presente proyecto utiliza el paradigma de la programación orientada a objetos para crear un programa que hace calculos sobre las siguientes figuras geometricas
# cuadrado, rectangulo, triángulo y circulo, para estas figuras dadas las medidas correspondientes se calcula su área, en adición a esto, al circulo se le calcula tambien el perimetro.
# Los calculos realizados se guardan en un contenedor que al final de la ejecución dice su area en base al total de las figuras calculadas

# área de un círculo
import math
from math import pi


class Circulo:
  def __init__(self, radio):
    self.radio = radio

  def get_radio(self):
    return self.radio

  def set_radio(self, x):
    self.radio = x

  def area_circulo(self):
    area = pi*self.radio**2
    return area

  def calcularPerimetro(self):
    perimetro = 2*pi*self.radio
    return perimetro

  def __str__(self):
    return "El circulo tiene un perímetro de " + str(self.calcularPerimetro())  + " y Area " + str(self.area_circulo()) + " en cm."


class triangulo:
  def __init__(self, base,altura):
    self.base = base
    self.altura = altura

  def get_base(self):
    return self.base

  def set_base(self, x):
    self.base = x

  def get_altura(self):
    return self.altura

  def set_altura(self, x):
    self.altura = x

  def area_triangulo(self,base,altura):
    area = (self.base*self.altura)/2
    return area

  def __str__(self):
    return "El circulo tiene un perímetro de " + str(self.calcularPerimetro())  + " y Area " + str(self.area_triangulo()) + " en cm."


class Rectangulo:

    def __init__(self, largo, ancho):
        self.largo = largo
        self.ancho = ancho

    def get_largo(self):
      return self.largo

    def get_ancho(self):
      return self.ancho

    def set_largo(self, largo):
      self.largo = largo

    def set_ancho(self, ancho):
      self.ancho = ancho

    def area_rectangulo(self):
        return self.largo * self.ancho

    def __str__(self):
        return f"El área del rectángulo es {self.area_rectangulo()}"


class Cuadrado:

    def __init__(self, lado):
        self.lado = lado


    def get_lado(self):
      return self.lado

    def set_lado(self, lado):
      self.lado = lado

    def area_cuadrado(self):
        return self.lado ** 2

    def __str__(self):
        return f"El área del cuadrado es {self.area_cuadrado()}"



class Figura:

  def __init__(self, tipo, area):
    self.tipo = tipo
    self.area = area

  def get_tipo(self):
    return self.tipo

  def set_tipo(self, x):
    self.tipo = x

  def get_area(self):
    return self.area

  def set_area(self, x):
    self.area = x


class Contenedor:

  def __init__(self):
    self.figuras = []

  def agregar_figura(self, figura):
    self.figuras.append(figura)

  def calcular_area_total(self):
    area_total = 0
    for figura in self.figuras:
      area_total += figura.get_area()
    return area_total

  def mostrar_figuras(self):
    for figura in self.figuras:
      return(f"{figura.get_tipo()}: {figura.get_area()}")

  def __str__(self):
    return f"El área total del contenedor es {self.calcular_area_total()} & la es figura es {self.mostrar_figuras()}"

def mostarMenu ():
  activo = True
  opcion = 0
  contenedor = Contenedor()
  print("*** MENÚ – CÁLCULOS SOBRE FIGURAS GEOMÉTRICAS ***\n")
  print(" 1. Área de un cuadrado\n")
  print(" 2. Área de un rectángulo\n")
  print(" 3. Área de un triángulo\n")
  print(" 4. Área de un círculo)\n")
  print(" 5. Perímetro de una circunferencia\n")

  cont = Contenedor()
  while True:

    try:
       opcion=int(input("\nPor favor Ingrese la opción deseada (1-5) o 0 para terminar: "))
       if opcion == 1:

         lado = int(input("\nPor favor Ingrese el lado del cuadrado \t"))
         print(lado)
         c1 = Cuadrado(lado)
         c2 = c1.area_cuadrado()
         print("\nEl área del cuadrado es", c2)
         fig = Figura("cuadrado", c2)
         cont.agregar_figura(fig)
         print(cont)

       elif opcion == 2:

        largo = int(input("\n Por favor Ingresa el largo del rectangulo \t"))
        ancho = int(input("\n Por favor Ingresa el ancho del rectangulo \t"))
        print(f'el largo del rectangulo es {largo} y el ancho es {ancho}')
        r1 = Rectangulo(largo, ancho)
        a = r1.area_rectangulo()
        print("\n El área del rectangulo es", a)
        fig = Figura("Area rectangulo ", a)
        cont.agregar_figura(fig)
        print(cont)

       elif opcion == 3:

        base = int(input("\n Por favor Ingresa la altura del triangulo \t"))
        altura = int(input("\n Por favor Ingresa la base del triangulo \t"))
        print(f'La base del triangulo es {base} y la altura es {altura}')
        t1 = triangulo(base, altura)
        a = t1.area_triangulo(base,altura)
        print("\n El área del triangulo es", a)
        fig = Figura("Area Triangulo ", a)
        cont.agregar_figura(fig)
        print(cont)

       elif opcion == 4:

        radio = int(input("\n Por favor Ingresa el radio del circulo \t"))
        print(radio)
        c1 = Circulo(radio)
        a = c1.area_circulo()
        print("\n El área del circulo es", a)
        fig = Figura("Area Circulo ", a)
        cont.agregar_figura(fig)
        print(cont)

       elif opcion == 5:

        radio = int(input("\n Por favor Ingresa el radio del circulo \t"))
        print(radio)
        c1 = Circulo(radio)
        p1 = c1.calcularPerimetro()
        print("\n El perimetro de la circunferencia es ", p1)
        fig = Figura("Perimetro circunferencia ", p1)
        cont.agregar_figura(fig)
        print(cont)

       elif opcion == 0:
        print(cont.mostrar_figuras())
        print(cont.calcular_area_total())
        break
    except:
       print("\nError")
  return opcion

mostarMenu ()
```
